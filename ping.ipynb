```python
import matplotlib.pyplot as plt
import re
import os

def read_ping_data(filename):
    sequences = []
    latencies = []
    with open(filename, 'r') as file:
        for line in file:
            match = re.search(r'icmp_seq=(\d+).*time=([\d.]+) ms', line)
            if match:
                sequences.append(int(match.group(1)))
                latencies.append(float(match.group(2)))
    return sequences, latencies

def plot_ping_data(sequences, latencies):
    plt.figure(figsize=(10, 6))
    plt.plot(sequences, latencies, 'b-', linewidth=2)  # Changed to a solid line without dots
    plt.xlabel('Time elapsed')
    plt.ylabel('Latency (ms)')
    plt.title('Ping Latency over Time')
    plt.grid(True)
    plt.tight_layout()

# Read data from ping.txt
sequences, latencies = read_ping_data('pingnuc.txt')

# Plot the data
plot_ping_data(sequences, latencies)

# Save the plot as an image file in the Downloads folder
downloads_path = os.path.expanduser("~/Downloads")
file_path = os.path.join(downloads_path, "ping.png")
plt.savefig(file_path, dpi=600, bbox_inches='tight')

# Display the plot (optional)
plt.show()

# Close the plot to free up memory
plt.close()
```
